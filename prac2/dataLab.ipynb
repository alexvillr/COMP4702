```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
```

# import dataset and show information

```python
df = pd.read_csv("./seedlings.csv")
```

```python
df.isnull().sum()
```

```python
dfDropped = df.dropna()
```

```python
dfDropped.isnull().sum()
```

```python
imputer = SimpleImputer(missing_values=pd.NA, strategy='mean')
dfImputed = df.drop(columns='Spec')
imputer.fit(dfImputed)
dfImputed = pd.DataFrame(imputer.transform(dfImputed))
print(dfImputed.info())
dfImputed.to_csv("./imputed_seedlings.csv")
```

```python
sb.pairplot(dfDropped[['DM_stem', 'DM_total', 'DM_roots', 'Shock', 'Leaf Area']])
sb.pairplot(dfImputed[[12, 14, 13, 15, 8]])
sb.pairplot(df[['DM_stem', 'DM_total', 'DM_roots', 'Shock', 'Leaf Area']])
```
